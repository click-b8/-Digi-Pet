```python
# Emoji Tamagotchi â€“ Animated TUI with Colored Status Bars
# Works in Colab or terminal that supports ANSI colors
# Run cell, then follow prompts

import time
import random
import sys

# -----------------------------
# ANSI colors
# -----------------------------
GREEN = "\033[92m"
YELLOW = "\033[93m"
RED = "\033[91m"
CYAN = "\033[96m"
RESET = "\033[0m"
BOLD = "\033[1m"

# -----------------------------
# Utility functions
# -----------------------------
def color_for(value):
    if value >= 70:
        return GREEN
    elif value >= 40:
        return YELLOW
    else:
        return RED
```

```python
3
233333111
``1


def bar(value, max_value=100, length=20):
    filled = int(length * value / max_value)
    color = color_for(value)
    return color + "â–ˆ" * filled + RESET + "â–‘" * (length - filled)


def bar_reverse(value, max_value=100, length=20):
    # Used for hunger: full bar = satisfied, empty = starving
    filled = int(length * (max_value - value) / max_value)
    color = color_for(max_value - value)
    return color + "â–ˆ" * filled + RESET + "â–‘" * (length - filled)


spinner_frames = ["â ‹","â ™","â ¹","â ¸","â ¼","â ´","â ¦","â §","â ‡","â "]

# -----------------------------
# Pet selection
# -----------------------------
pets = {
    "1": ("ðŸ£", "Chick"),
    "2": ("ðŸ±", "Cat"),
    "3": ("ðŸ¶", "Dog"),
    "4": ("ðŸ¸", "Frog"),
    "5": ("ðŸ‘¾", "Alien"),
    "6": ("ðŸ²", "Dragon"),
    "7": ("ðŸ¦Š", "Fox"),
    "8": ("ðŸ¼", "Panda")
}

print(BOLD + "Choose your digital pet:" + RESET)
for k, v in pets.items():
    print(f"{k}. {v[0]}  {v[1]}")

choice = input("Enter number: ")
pet_emoji, pet_name = pets.get(choice, pets["1"])

# -----------------------------
# Initial stats
# -----------------------------
hunger = 40
happiness = 60
energy = 60
age = 0
alive = True
spin = 0

# -----------------------------
# Main display
# -----------------------------
def draw():
    global spin
    print("\n" * 50)

    if not alive:
        print(RED + "ðŸ’€ Your pet has run away..." + RESET)
        sys.exit()

    mood = "ðŸ™‚"
    if hunger > 80:
        mood = "ðŸ˜¡"
    elif happiness < 30:
        mood = "ðŸ˜¢"
    elif energy < 30:
        mood = "ðŸ¥±"

    spinner = spinner_frames[spin % len(spinner_frames)]
    spin += 1

    print(BOLD + "==============================" + RESET)
    print(f"   {pet_emoji}  {mood}   {CYAN}{spinner}{RESET}")
    print(f"   {pet_name}   Age: {age}")
    print(BOLD + "==============================" + RESET)

    print(f"Hunger    [{bar_reverse(hunger)}] {hunger}/100")
    print(f"Happiness [{bar(happiness)}] {happiness}/100")
    print(f"Energy    [{bar(energy)}] {energy}/100")

    print(BOLD + "==============================" + RESET)
    print("Actions:")
    print("1. Feed ðŸŽ")
    print("2. Play ðŸŽ®")
    print("3. Sleep ðŸ˜´")
    print("4. Clean ðŸ§¼")
    print("5. Do nothing")

# -----------------------------
# Game tick
# -----------------------------
def tick():
    global hunger, happiness, energy, age, alive

    hunger = min(100, hunger + random.randint(3, 6))
    happiness = max(0, happiness - random.randint(2, 4))
    energy = max(0, energy - random.randint(1, 3))
    age += 1

    if hunger >= 100 or happiness <= 0:
        alive = False

# -----------------------------
# Actions
# -----------------------------
def feed():
    global hunger, happiness
    hunger = max(0, hunger - 25)
    happiness = min(100, happiness + 5)


def play():
    global happiness, energy
    happiness = min(100, happiness + 18)
    energy = max(0, energy - 12)


def sleep():
    global energy, hunger
    energy = min(100, energy + 30)
    hunger = min(100, hunger + 8)


def clean():
    global happiness
    happiness = min(100, happiness + 8)

# -----------------------------
# Main loop
# -----------------------------
while alive:
    draw()
    action = input("Choose action: ")

    if action == "1":
        feed()
    elif action == "2":
        play()
    elif action == "3":
        sleep()
    elif action == "4":
        clean()

    tick()
    time.sleep(0.4)
```
